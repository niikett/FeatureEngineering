```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
np.random.seed(42)
data = {
    "age": np.random.randint(25, 76, 100),
    "salary": np.random.randint(25000, 1000000, 100),
    "purchased": np.random.choice([0, 1], size=100)
}
```

```python
df = pd.DataFrame(data)
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.iloc[:, 0:-1], df.iloc[:, -1], test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()

scaler.fit(x_train)

x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)

# it returns the numpy array

# converting in the data frame

x_train_scaled = pd.DataFrame(x_train_scaled, columns=x_train.columns)
x_test_scaled = pd.DataFrame(x_test_scaled, columns=x_test.columns)
```

```python
print("non scaled:")
print(x_train.describe().loc[["mean", "std"]])

print("\n")

print("scaled:")
print(x_train_scaled.describe().loc[["mean", "std"]])
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

axes[0].scatter(x_train["salary"], x_train["age"])
axes[0].set_title("non scaled")

axes[1].scatter(x_train_scaled["salary"], x_train_scaled["age"])
axes[1].set_title("scaled")
```
