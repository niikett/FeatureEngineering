```python
import random
import string
import numpy as np
import pandas as pd
```

```python
np.random.seed(42)

data = {
    "x": [random.choice([random.choice(string.ascii_lowercase), str(np.random.randint(1, 100))]) for _ in range(100)],
    "y": [random.choice(string.ascii_lowercase) + str(np.random.randint(1, 100)) for _ in range(100)]
}
```

```python
df = pd.DataFrame(data)
```

##### data type : [numeric, categorical]

```python
df['x_numerical'] = pd.to_numeric(df["x"], errors="coerce")
df["x_categorical"] = np.where(df["x_numerical"].isnull(), df["x"], np.nan)

df
```

##### data type : [numeric+categorical, numeric+categorical]

```python
df["y_numerical"] = df["y"].str.extract("(\d+)")
df["y_categorical"] = df["y"].str[0]

df
```
