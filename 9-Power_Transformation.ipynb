```python
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as stats
import matplotlib.pyplot as plt
from sklearn.preprocessing import PowerTransformer
```

#### box cox transformation

```python
np.random.seed(42)

data = {
    "positive feature": np.random.exponential(scale=2, size=100)      
}
```

```python
df = pd.DataFrame(data)
```

```python
boxcox_transformer = PowerTransformer(method="box-cox")
```

```python
df["positive feature BoxCox"] = boxcox_transformer.fit_transform(df[["positive feature"]])
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1,2,1)
sns.kdeplot(df["positive feature"], fill=True)  
plt.xlabel('x values')
plt.ylabel('Density')
plt.title('PDF curve for x')

# plt.subplot(2,2,2)
# stats.probplot(df["x"], dist="norm", plot=plt)
# plt.title("Q-Q Plot of x")

plt.subplot(1,2,2)
sns.kdeplot(df["positive feature BoxCox"], fill=True)  
plt.xlabel('x_transform values')
plt.ylabel('Density')
plt.title('PDF curve for x_transform')

# plt.subplot(2,2,4)
# stats.probplot(x_transform, dist="norm", plot=plt)
# plt.title("Q-Q Plot of x_transform")

plt.show()
```

#### yeo johnson transformation

```python
np.random.seed(42)

data = {
    "mixed feature": np.random.uniform(-5, 5, size=100) ** 3    
}
```

```python
df = pd.DataFrame(data)
```

```python
yeojohnson_transformer = PowerTransformer(method="yeo-johnson")
```

```python
df["mixed feature YeoJohnson"] = yeojohnson_transformer.fit_transform(df[["mixed feature"]])
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1,2,1)
sns.kdeplot(df["mixed feature"], fill=True)  
plt.xlabel('x values')
plt.ylabel('Density')
plt.title('PDF curve for x')

# plt.subplot(2,2,2)
# stats.probplot(df["x"], dist="norm", plot=plt)
# plt.title("Q-Q Plot of x")

plt.subplot(1,2,2)
sns.kdeplot(df["mixed feature YeoJohnson"], fill=True)  
plt.xlabel('x_transform values')
plt.ylabel('Density')
plt.title('PDF curve for x_transform')

# plt.subplot(2,2,4)
# stats.probplot(x_transform, dist="norm", plot=plt)
# plt.title("Q-Q Plot of x_transform")

plt.show()
```
