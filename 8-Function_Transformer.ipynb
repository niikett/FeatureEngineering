```python
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as stats
import matplotlib.pyplot as plt
from sklearn.preprocessing import FunctionTransformer
```

#### log transformation

```python
np.random.seed = 42

data = {
    "x": np.power(10, np.arange(10))
}
```

```python
df = pd.DataFrame(data)
```

```python
trf = FunctionTransformer(np.log1p)

x_transform = trf.transform(df["x"])
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1,2,1)
sns.kdeplot(df["x"], fill=True)  
plt.xlabel('x values')
plt.ylabel('Density')
plt.title('PDF curve for x')

# plt.subplot(2,2,2)
# stats.probplot(df["x"], dist="norm", plot=plt)
# plt.title("Q-Q Plot of x")

plt.subplot(1,2,2)
sns.kdeplot(x_transform, fill=True)  
plt.xlabel('x_transform values')
plt.ylabel('Density')
plt.title('PDF curve for x_transform')

# plt.subplot(2,2,4)
# stats.probplot(x_transform, dist="norm", plot=plt)
# plt.title("Q-Q Plot of x_transform")

plt.show()
```
