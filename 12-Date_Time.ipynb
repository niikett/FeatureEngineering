```python
import numpy as np
import pandas as pd
```

```python
def generate_random_dates():
    years = np.random.randint(2000, 2015, 100)
    months = np.random.randint(1, 12, 100)
    days = np.random.randint(1, 28, 100)

    dates = [f"{y:04d}-{m:02d}-{d:02d}" for y, m, d in zip(years, months, days)]

    return dates

def generate_random_times():
    hours = np.random.randint(0, 23, 100)
    minutes = np.random.randint(0, 59, 100)
    seconds = np.random.randint(0, 59, 100)
    
    times = [f"{h:02d}:{m:02d}:{s:02d}" for h, m, s in zip(hours, minutes, seconds)]

    return np.array(times)

np.random.seed(42)

data = {
    "date": generate_random_dates(),
    "time": generate_random_times()
}
```

```python
df = pd.DataFrame(data)
```

```python
df.info()
```

```python
df["date"] = pd.to_datetime(df["date"])
df["time"] = pd.to_datetime(df["time"], format="%H:%M:%S")
```

```python
df.info()
```

```python
df["date_year"] = df["date"].dt.year
df["date_month"] = df["date"].dt.month
df["date_month_name"] = df["date"].dt.month_name()
df["date_day"] = df["date"].dt.day
df["date_day_name"] = df["date"].dt.day_name()


df["time_hour"] = df["time"].dt.hour
df["time_minute"] = df["time"].dt.minute
df["time_second"] = df["time"].dt.second
```
